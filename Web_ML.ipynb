```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
df = pd.read_csv("austin_final.csv")
df[df['Events']== 'Rain']
h= df.groupby('Events')
h.mean()
```

```python
features = ['DewPointHighF','DewPointAvgF','DewPointLowF','HumidityHighPercent','HumidityAvgPercent','HumidityLowPercent','WindHighMPH','WindGustMPH']

target = 'PrecipitationSumInches'

x = df[features].values
y = df[target].values

x_mean = np.mean(x, axis=0)
x_sd = np.std(x, axis = 0)
x_normalized = (x - x_mean) / x_sd

df
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x,y)

y_pred = model.predict(x)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y, y_pred)

r2 = r2_score(y, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
